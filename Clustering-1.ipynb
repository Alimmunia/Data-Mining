```python
from sklearn import datasets
from sklearn.datasets import make_blobs
X,y = make_blobs(n_samples=100,centers=4,random_state=0,cluster_std=0.60)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1], s=50)
```

```python
from sklearn.cluster import KMeans
est = KMeans(4)
est.fit(X)
y_kmeans = est.predict(X)
plt.scatter(X[:,0],X[:,1],c=y_kmeans,s=50,cmap='rainbow')
```

```python
iris = datasets.load_iris()

X,y = iris.data,iris.target
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X)
```

```python
X_reduced = pca.transform(X)
print("Bentuk dataset setelah direduksi:",X_reduced.shape)
```

```python
import pylab as pl
pl.scatter(X_reduced[:,0],X_reduced[:,1],c=y,cmap='rainbow')
```

```python
for component in pca.components_:
  print("+".join("%.3f x %s" % (value, name)
    for value, name in zip(component, iris.feature_names)))
```

```python
k_means = KMeans(n_clusters=3,random_state=0)
k_means.fit(X)
y_pred = k_means.predict(X)
```

```python
pl.scatter(X_reduced[:,0],X_reduced[:,1],c=y_pred,cmap='rainbow');
```
